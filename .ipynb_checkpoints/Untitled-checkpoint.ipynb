checking number if number of unemployed people grows with age
#third important attribute , year

```python
#copy sorted df_profiles
df_profiles_sorted=df_profiles[df_profiles['job'].isnull()].sort_values(by='birthdate')
#keep only yaer
df_profiles_sorted['birthdate']=pd.to_datetime(df_profiles_sorted['birthdate']).dt.year
#create graph
plt.plot(df_profiles_sorted.groupby(by=['birthdate'])['birthdate'].count())
#make a funny line with linegress since idk what to do 
res = linregress(df_profiles_sorted.groupby(by=['birthdate'])['birthdate'].count().index, df_profiles_sorted.groupby('birthdate')['birthdate'].count())
plt.plot(df_profiles_sorted['birthdate'], res.intercept + res.slope * df_profiles_sorted['birthdate'])
```
